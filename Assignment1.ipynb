###3.1 Artificial Neural Networks (ANN)
With the help of neurons, which are the four components listed as input data, weights, transfer function and output data. These components build a network which have three layers i.e. input layer, hidden layer and output layer. Artificial neural network is used to perform dynamic tasks which are non-linear in system. It produces a single output on the basis of multiple input, and the result is achieved by comparing the output produced with the desired output. Hence, the transfer between the neurons takes place until the output is close to the desired output. ANN is divided in two main types, the first one is forward feed network, second is recurrent network which performs recursive algorithm. At the beginning random weights and bias are assigned, after calculating the error signals and desire to reach the original output, which is done by back propagation of weights through transfer function (Abbas et al. 2019).

###3.2 Support Vector Machine (SVM)
Support vector machine lays out a high-dimensional hyperplane H plus and H minus on which the parameters are separated optimally in a non-linear fashion. For optimizing parameters, generalization and regularization is needed. In this study, kernel functions are used to reduce model errors and complexity of feature space, for which Gaussian and Polynomial will be applied on the parameters. To create a boundary between two classes that divides the hyperplane are H+ = w.p – b = +1 and H- = w.p - = -1, which is done by classification and regression technique. With the help of Lagrangian it Is possible to optimize the feature space in a dual space for binary classification (Abbas et al. 2019).
